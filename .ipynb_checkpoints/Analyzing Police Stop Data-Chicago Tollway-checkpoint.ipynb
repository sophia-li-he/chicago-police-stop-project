```python
import os
os_cwd = os.getcwd()
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import warnings
warnings.simplefilter('ignore')
```

```python
def singlevalue(my_series):
    """
    input: pd.series
    output: the number of single values
    """
    value = len(my_series.value_counts()[my_series.value_counts() == 1])
    return value
def df_explore(data_explore):
    """
    input: pandas dataframe
    output: 
    1. the shape of the data frame
    2. the number of unique value in each column
    3. the percentage of null value in each column
    4. the number of single value in each column
    """
    print ("1. The shape of the data is " + str(data_explore.shape))
    data_explore_unique = pd.DataFrame(data_explore.apply(lambda x: x.nunique(), axis = 0))
    data_explore_unique.rename(columns={0:'nunique'}, inplace=True)
    data_explore_null = pd.DataFrame(data_explore.isnull().sum()/len(data_explore)*100)
    data_explore_null.rename(columns={0:'null'}, inplace=True)
    data_explore_single = pd.DataFrame(data_explore.apply(lambda x: singlevalue(x), axis = 0))
    data_explore_single.rename(columns={0:'nsingle'}, inplace=True)
    data_explore_result = data_explore_unique.join(data_explore_null, how='outer').join(data_explore_single, how='outer')
    return data_explore_result
```

```python
il_original = pd.read_csv('IL-clean.csv')
il = il_original.copy()
il = il.loc[il.location_raw == 'ILLINOIS STATE POLICE 15', ]
df_explore_il = df_explore(il)
df_explore_il
```

# How likely will you get a ticket on Chicago tollways?<a class="anchor" id="7"></a>
[Back to TOC](#toc)

## Feature Engineering
1. Categorical variables: `driver_gender`, `driver_race`, `violation`, `stop_datetime(hour_group)`
2. Boolean variables: `search_conducted`, `contraband_found`
3. Numeric variables: `driver_age`
4. Dependent variable: if a ticket (citation) is issued

```python
# Baseline prediction accuracy
404004 / (404004 + 319047)
```

```python
il['stop_hour'] = il.stop_datetime.dt.hour
il.loc[il.stop_hour.isin([22, 23, 0, 1, 2, 3, 4]), 'stop_hour_group'] = '22-4'
il.loc[~il.stop_hour.isin([22, 23, 0, 1, 2, 3, 4]), 'stop_hour_group'] = '5-21'
il['stop_hour_group'].value_counts()
```

### Transform categorical variables

```python
il_categorical_col = ['driver_gender', 'driver_race', 'violation', 'stop_hour_group']
```

### Drop null rows

```python
il = il.dropna(subset = ['driver_age', 'stop_datetime'])
```

```python
def create_dummy_df(df, cat_cols, dummy_na):
    cat_df = pd.get_dummies(df[cat_cols], prefix = cat_cols, prefix_sep='_', dummy_na = dummy_na, drop_first=False)
    df_copy = df.drop(cat_cols, axis = 1)
    df_copy = df_copy.merge(cat_df, how = 'inner', left_index = True, right_index = True)
    return df_copy
```

```python
il_input = il[['driver_gender', 'driver_race', 'violation', 'stop_hour_group', 'driver_age', 'search_conducted', 'contraband_found']]
il_X = create_dummy_df(il_input, il_categorical_col, dummy_na = False)
il_y = il['ticket']
```

```python
# Make sure the input data is as we expected
df_explore(il_X)
```

## Logistic Regression Model

```python
logreg = linear_model.LogisticRegression()
X_train, X_test, y_train, y_test = train_test_split(il_X, il_y, test_size = .25, random_state=42) 
logreg.fit(X_train, y_train)
train_pred = logreg.predict(X_train)
y_pred = logreg.predict(X_test)
accuracy_train = np.mean(train_pred == y_train)
accuracy = np.mean(y_pred == y_test)
print("train accuracy = ", accuracy_train * 100, "%")
print("accuracy = ", accuracy * 100, "%")
```

```python
coeff = list(logreg.coef_[0])
labels = list(X_train.columns)
features = pd.DataFrame()
features['Features'] = labels
features['importance'] = coeff
features.sort_values(by=['importance'], ascending=True, inplace=True)
features['positive'] = features['importance'] > 0
features.set_index('Features', inplace=True)
features.importance.plot(kind='barh', figsize=(20, 10),color = features.positive.map({True: 'blue', False: 'red'}))
plt.xlabel('Importance')
#plt.savefig('coef.png')
```
